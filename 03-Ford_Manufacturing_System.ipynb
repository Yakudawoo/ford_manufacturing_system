## Task 1 - Car Class

Every car that rolls off the Ford assembly line has specific characteristics. Let's create a Car class to represent these vehicles.

1. Create a class called `Car`.
2. The `__init__` method should take the following parameters:
   - `model` (e.g., "Mustang", "F-150", "Explorer")
   - `color` (e.g., "Red", "Blue", "Black")
   - `year` (e.g., 2024)
   - `price` (e.g., 35000)
   
   Store each parameter as an attribute using `self.attribute_name = parameter_name`

3. Create a method `get_info()` that returns a formatted string with all the car's information.

<Note type="note">
    
       Use an f-string to format the output nicely, for example: f"{self.year} {self.color} {self.model} - ${self.price}"
</Note>

4. Create a method `apply_discount(percentage)` that reduces the car's price by the given percentage.
   - Calculate the new price using the formula below
   - Update `self.price` with the new value
   - Print a message showing the new price

    To calculate a discount: `new_price = price - (price * percentage / 100)`

```python
class Car:
    def __init__(self, model, color, year, price):
        self.model = model
        self.color = color
        self.year = year
        self.price = price
    
    def get_info(self):
        return f"{self.year} {self.color} {self.model} - ${self.price}"
    
    def apply_discount(self, percentage):
        new_price = self.price - (self.price * percentage / 100)
        self.price = new_price
        print(f"New price after {percentage}% discount is: ${self.price}")
```

### Test your Car class

```python
# Create a car and display its info
mustang = Car("Mustang", "Red", 2024, 45000)
print(mustang.get_info())
mustang.apply_discount(10)
print(f"After discount: ${mustang.price}")
```

## Task 2 - The Inventory Class

Ford needs to keep track of all the cars in stock. Let's create an Inventory system to manage the available vehicles.

1. Create a class called `Inventory`.
2. In the `__init__` method, initialize an empty list called `cars` to store all Car objects.
3. Create a method `add_car(car)` that adds a Car object to the inventory.
4. Create a method `remove_car(model, color)` that removes the first car matching the model and color from inventory.
5. Create a method `get_total_value()` that calculates and returns the total value of all cars in inventory.
6. Create a method `find_cars_by_model(model)` that returns a list of all cars of a specific model. 

<Note type="hint">
    Don't forget to initialize a new list
</Note>

7. Create a method `display_inventory()` that prints information about all cars in stock.

<Note type="note">
Remember you can use built-in function such as `.appends()` or `.remove() in order to add or remove an item from a list.
</Note>

```python
class Inventory:
    def __init__(self): # ou self.cars = list()
        self.cars=[]
    
    def add_car(self, car):
        self.cars.append(car)
    
    def remove_car(self, model, color):
        car_to_remove = [car for car in self.cars if not (car.model == model and car.color == color)]
        self.cars.remove(car_to_remove)
        if car_to_remove:
            self.cars.remove(car_to_remove[0])
            return car_to_remove[0]
        return None
    
    def get_total_value(self):
        return sum(car.price for car in self.cars)
    
    def find_cars_by_model(self, model):
        return [car for car in self.cars if car.model == model]
    
    def display_inventory(self):
        if not self.cars:
            print("Inventory is empty.")
            return # ou break?
        for car in self.cars:
            print(f"Current Inventory:\n\n {car.get_info()}")
            
 
```
